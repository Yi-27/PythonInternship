```python
# 爬取实习网（
import requests
from lxml import etree

# 网址
url = 'https://www.shixi.com/'

# 设置请求头
header = {
    'User-Agent': 'User-Agent: Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/83.0.4103.116 Safari/537.36 Edg/83.0.478.61'
}

res = requests.get(url, headers=header, timeout=2.0) # 设置请求头和超时

# 打印状态码
print(res.status_code)

# 打印编码格式
print(res.encoding)

# 打印请求头
print(res.headers)

html = etree.HTML(res.content, etree.HTMLParser()) # 使用html解析器
```

```python
import pandas as pd
# 爬取实习网热门职位的相关信息（职位、地址、学历要求。工资）
# 根据题意这里应该只是要爬取首页显示的热门职业
jobs = []
companies = []
addresses = []
degrees = []
salaries = []

Chinese_items = html.xpath('//*[@class="job_list clearfix"]/li/div') # 提取出中国的热门职业
print("中国热门：", len(chinese_items))
for i in Chinese_items:
    job = i.xpath('./h3/a/text()') # 职位
    company = i.xpath('./p[1]/a/text()') # 公司名称
    address = i.xpath('./p[@class="job_detail"]/span[@class="address"]/text()') # 地址
    degree = i.xpath('./p[@class="job_detail"]/span[@class="degree"]/text()') # 学历要求
    salary = i.xpath('./p[@class="job_detail"]/span[@class="money"]/text()') # 工资
    print(job, company, address, degree, salary)
    jobs.append(job[0])
    companies.append(company[0])
    addresses.append(address[0])
    degrees.append(degree[0])
    salaries.append(salary[0])


America_items = html.xpath('//*[@class="job_list clearfix"]/li/a[@class="job_list_item"]/div') # 提取出中国的热门职业
print("美国热门：", len(America_items))
for i in America_items:
    job = i.xpath('./h3/span[1]/text()') # 职位
    company = i.xpath('./p[1]/text()') # 公司名称
    address = i.xpath('./p[@class="job_detail"]/span[@class="address"]/text()') # 地址
    degree = i.xpath('./p[@class="job_detail"]/span[@class="degree"]/text()') # 学历要求
    salary = i.xpath('./p[@class="job_detail"]/span[@class="money"]/text()') # 工资
    print(job, company, address, degree, salary)
    jobs.append(job[0])
    companies.append(company[0])
    addresses.append(address[0])
    degrees.append(degree[0])
    salaries.append(salary[0])
    
pd.DataFrame({"job": jobs, "conpany": companies, "address": addresses, "degree": degrees, "salary": salaries})
```
