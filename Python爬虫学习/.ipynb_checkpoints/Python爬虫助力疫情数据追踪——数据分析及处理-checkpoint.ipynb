## 分析处理国内历史每日数据

```python
# 查看国内历史每日数据
china_day_data = pd.read_csv(filepath_cn+"国内历史每日数据.csv")
china_day_data
```

```python
# 为了更直观看，将列名全部改为中文
# field_name_cn = {'name': '地区', 'nowConfirm': '现有确诊', 'confirm': '累计确诊', 'suspect': '疑似', 'dead': '死亡', 'deadRate': '死亡率', 'heal': '治愈', 'healRate': '治愈率'}
china_day_data.columns = ['累计确诊', '疑似', '累计死亡', '累计治愈', '现有确诊', '现有重症患者', '境外输入', '死亡率', '治愈率', '日期', '无症状感染者']
```

```python
# 日期列调到第一列
data_col = china_day_data.pop('日期') # 先弹出该列
china_day_data.insert(0, '日期', data_col) # 再将该列添加到第一列即可
```
