```python
# 完整的请求
import urllib3

# 生成一个实例
http = urllib3.PoolManager()

# 网址
url = 'http://www.tipdm.com/tipdm/index.html'

# 设置请求头
header = {
    'User-Agent': 'User-Agent: Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/83.0.4103.116 Safari/537.36 Edg/83.0.478.61'
}

# 超时时间
time_out = urllib3.Timeout(connect=1.0, read=3.0)

# 重试次数和重定向次数设置并生成请求
res = http.request('GET',
                    url=url, 
                    headers=header, # 添加请求头
                    timeout=time_out,
                    retries=5, # 设置重试次数
                    redirect=4 # 设置重定向次数
                  ) # 分别设置连接超时时间和读取超时时间
print("服务器响应码：", res.status) # 200表示成功
print("响应实体：", res.data.decode('utf-8')) # 结果与上面是一样的 解码的时候用utf-8解码,就可以正常显示中文了
```

# Python正则表达式

```python
import re

string = "1. A small sentence. - 2. Another tiny sentence. "

# 这三个如果匹配成功
print(re.findall('sentence', string)) # 匹配所有符合的, 返回匹配到所有构成的列表
print(re.search('sentence', string)) # 匹配一个符合的，返回对象，包括单词的起始和终止的索引位置
print(re.match('sentence', string)) # match默认从第一个字符开始往后找，若匹配到，会返回单词的起始和终止的索引位置
re.sub(pattern='small', repl='large', string=string) # pattern是匹配的规则（上面三个省略了） repl是替换的字符串（可以为空）
```

```python
string = "small smAll SmAll SmEll smell smll smsmll sm3ll sm.ll sm?ll sm-ll sm\nll sm\tll"

print(re.findall('sm.ll', string)) # . 匹配除换行符以外所有的字符
print(re.findall('sm.ll', string, re.S)) # re.S 使 . 匹配包括换行符在内的所有字符
print(re.findall('sm[asdbf]ll', string)) # 只匹配匹配[]内的有的字符
print(re.findall('sm[asdbf]ll', string, re.I)) # re.I 表示不区分大小写
print(re.findall('sm[a-zA-Z0-9]ll', string)) # [a-zA-Z0-9]匹配数字和大小写字母
print(re.findall('sm\.ll', string)) # 匹配 . 要进行转义
print(re.findall('sm[.?\-]ll', string)) # 但是在[]就不用转义了，如果要匹配 -  就需要转义了
print(re.findall('small|smell', string)) # 匹配small 或 smell
print(re.findall('sm\wll', string)) # 同 [a-zA-Z0-9]匹配数字和大小写字母
```

```python
print(re.findall('sm..ll', string)) # 匹配2个未知字符（除换行符）
print(re.findall('sm.{2}ll', string)) # 表示前面的 . 的效果要匹配两次，同上面
print(re.findall('sm.{1,2}ll', string)) # 匹配至少匹配1次，至多匹配2次 注意{1,2}中间不能有空格
print(re.findall('sm.{1,}ll', string)) # 匹配至少匹配1次
print(re.findall('sm.?ll', string)) # ? 表示前面的元素被匹配1次或0次

# 这里是正则表达式的贪婪匹配，所以匹配到整个个字符串； 
# 但是由于 . 不能匹配换行符，所以sm\nll没被匹配到，但是sm\tll继续又被匹配到了
print(re.findall('sm.+ll', string)) # 匹配前面的元素被匹配1次或多次
print(re.findall('sm.*ll', string)) # 匹配前面的元素被匹配0次或多次
```

```python
# 小练习
"""
555-1239Moe Szyslak
(636) 555-0113Burns, C.Montgomery
555-6542Rev. Timothy Lovejoy
555 8904Ned Flanders
636-555-3226Simpson,Homer
5553642Dr. Julius Hibbert
"""
import pandas as pd
rawdata = "555-1239Moe Szyslak(636) 555-0113Burns, C.Montgomery555-6542Rev. Timothy Lovejoy555 8904Ned Flanders636-555-3226Simpson,Homer5553642Dr. Julius Hibbert"

names = re.findall('[A-Z][A-Za-z,. ]*', rawdata) # 匹配所有名字，这里英文名字由大小写、点、空格、逗号构成 这里添加了*，所以是贪婪匹配
print(names)

phones = re.findall('\(?[0-9]{0,3}\)?[ \-]?[0-9]{3}[ \-]?[0-9]{4}', rawdata) # ()在正则中是提取，所以这里要用转义符
print(phones) # 同时因为() 和 - 不一都有，所以加 ? 表示匹配1次或0次 另外 中间[ \-] 里面还要匹配一个空格

pd.DataFrame({'Names': names, 'TelPhone': phones})
```

# 使用正则表达式获取网页标题内容

```python
import requests
import re

# 网址
url = 'http://www.tipdm.com/tipdm/index.html'

# 设置请求头
header = {
    'User-Agent': 'User-Agent: Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/83.0.4103.116 Safari/537.36 Edg/83.0.478.61'
}

res = requests.get(url, headers=header, timeout=2.0) # 设置请求头和超时
res.encoding='utf-8'
res.text
```

```python
# 提取标题
re.findall('<li><a href=".*" target="">(.*)</a></li>', res.text) # 直接 。*表示随便页面地址是啥
```

```python
#  <li class=" on"><a href="/">首页</a></li>
print(re.findall('<li[ a-z="]*><a.*href=".*".*>(.*)</a></li>', res.text)) # 直接 。*更粗暴，但是这里会匹配到所有符合正则规则的
# 如果想只提取到标题内容的话，可以先提取导航栏，再提取标题
# nav = re.findall('<nav class="nav">(.*)</nav>', res.text) # 这样无法直接提取出来
start = re.search('<nav class="nav">', res.text) # 但是可以通过找到中间内容的位置来通过字符串切片提取出中间的内容
print(start)
end = re.search('</nav>', res.text)
print(end)

# 那么内容就在 span=(1739-2351)的范围内了
nav = res.text[1739:2351] # 通过看上面的res.text发现，这里面开头有换行符，所以打印的打印的时候有空行
print("------------------------")
print(nav)
print("------------------------")
print(re.findall('<li[ a-z="]*><a.*href=".*".*>(.*)</a></li>', nav)) # 再提取一次，就只剩标题了
```

# 使用XPath解析网页

```python
import requests
from lxml import etree

# 网址
url = 'http://www.tipdm.com/tipdm/index.html'

# 设置请求头
header = {
    'User-Agent': 'User-Agent: Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/83.0.4103.116 Safari/537.36 Edg/83.0.478.61'
}

res = requests.get(url, headers=header, timeout=2.0) # 设置请求头和超时
html = etree.HTML(res.content, etree.HTMLParser()) # 使用html解析器
html
```

```python
etree.tostring(html, encoding='utf-8').decode('utf-8') # 字节型的字符串才能使用decode方法
# 这样的效果和res.text一样的
```

```python
# 提取标题
html.xpath('/html/head/title')
```

```python
html.xpath('//title')
```

```python
print(html.xpath('/html/body/header/div/nav/ul/li')) # 提取导航栏元素
print(html.xpath('/html/body/header/div/nav/ul/li/a/text()')) # 提取出具体的内容
print(html.xpath('//ul[@id="menu"]/li/a/text()')) # 简短写
```

```python
print(html.xpath('/html/body/header/div/nav/ul/li[1]')) # 提取第一个
print(html.xpath('/html/body/header/div/nav/ul/li[1]/a/text()'))
```

```python
print(html.xpath('/html/body/header/div/nav/ul/li[last()]')) # 提取最后一个
print(html.xpath('/html/body/header/div/nav/ul/li[last()]/a/text()'))
```

# 数据存储

```python
import requests
from bs4 import BeautifulSoup

# 网址
url = 'http://www.tipdm.com/tipdm/index.html'

# 设置请求头
header = {
    'User-Agent': 'User-Agent: Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/83.0.4103.116 Safari/537.36 Edg/83.0.478.61'
}

res = requests.get(url, headers=header, timeout=2.0) # 设置请求头和超时

soup = BeautifulSoup(res.content, 'lxml') # 解析网页
data = soup.select('#menu > li > a')
names = [i.text for i in data]
href = [i['href'] for i in data]
print(names)
href
```

```python
import json

# 保存为json文件
with open('./temp.json', 'w') as f:
    json.dump({'names': names, 'href': href}, f, ensure_ascii=False) # ensure_ascii=False中文正常显示

# 读取json文件
with open('./temp.json', 'r') as f:
    json_data = json.load(f)
print(json_data)
# 除了dump()和load()读写json文件，还有
# loads() :将json字符串转换成字典格式;dumps() 将字典格式数据转换成json格式;
```

```python
from sqlalchemy import create_engine
import pandas as pd

engine = create_engine("mysql+pymysql://root:rootroot@127.0.0.1:3306/test?charset=utf8")

print(pd.DataFrame({'names': names, 'href': href}))
pd.DataFrame({'names': names, 'href': href}).to_sql('tempjson', engine) # 保存到数据库中
```

```python
# 再从数据库中读取出来
pd.read_sql('tempjson', con=engine)
```

```python
from selenium import webdriver
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.common.by import By

# 由于默认的Edge的webdriver和我下的不太一样，所以这里要指定路径
driver_path = "F:/Python/msedgedriver.exe"
driver = webdriver.Edge(executable_path=driver_path) # 指定driver路径

driver.get("https://ptpress.com.cn/shopping/index") # 浏览器打开页面

wait = WebDriverWait(driver, 10) # 显式等待10秒

before = driver.current_window_handle # 记录当前窗口

search_btn = driver.find_element_by_css_selector(
    '#header > div.container > div > div.col-md-8.tools > div.search > input[type=text]'
)
search_btn.send_keys('Python编程') # 输入文本

confirm_btn = wait.until(EC.element_to_be_clickable(
    (By.CSS_SELECTOR, '#header > div.container > div > div.col-md-8.tools > div.search > button')
)) # 直到元素可被点击

confirm_btn.click() # 点击搜索
```

```python
driver.switch_to.window(driver.window_handles[1]) # 移动窗口到新打开的搜索页 switch_to.window这个是推荐用法
# driver.window_handles[1] 表示切换到第二个标签页上
html = driver.page_source # 这时page_source就是新的页面的源代码了
html
```

```python
from bs4 import BeautifulSoup

soup = BeautifulSoup(html, 'lxml')
book_id = [i['href'][22:] for i in soup.select('#books_content > div > div.book_item > a')]
img_src = [i['src'] for i in soup.select('#books_content > div > div.book_item > a > div > img')]
book_name = [i.text for i in soup.select('#books_content > div > div.book_item > a > p')]
print(book_id)
print(img_src)
book_name
```

```python
import pandas as pd
pd.DataFrame({'书名': book_name, '书编号': book_id, '书封面地址': img_src})
```
