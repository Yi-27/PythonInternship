```python
import requests
import matplotlib.pyplot as plt

s = requests.Session() # 创建会话

url = 'https://ptpress.com.cn/login'

res = s.get("https://ptpress.com.cn/kaptcha.jpg") # 这里的get是在会话里发生的
with open('./kaptcha.jpg', 'wb') as f:
    f.write(res.content) # 保存图片到本地
    
pic = plt.imread('./kaptcha.jpg') # 读取图片
plt.imshow(pic) # 显示图片
plt.show()

a = input("请输入验证码：")

# 表单
login = {
    'username': '17355477371', # 这里并没有用真实的用户名和密码
    'password': '123456',
    'verifyCode': a
}
rqq = s.post(url, data=login) # 发送post请求，模拟登陆
print(rqq.content.decode('utf-8'))
```
