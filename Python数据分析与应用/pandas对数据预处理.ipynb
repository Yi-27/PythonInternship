### 插值法

```python
x = np.array([1, 2, 3, 6, 7])
y = np.array([3, 5, 6, 9, 13])

from scipy.interpolate import interp1d
model = interp1d(x, y, kind='linear') # 使用线性插值
```

```python
model([4, 5])
```

```python
import matplotlib.pyplot as plt
plt.scatter(x, y) # 绘制散点图
plt.plot(x, model(x), 'r-') # 绘制插值后的拟合的线
```

```python
from scipy.interpolate import lagrange
f_lag = lagrange(x, y) # 拉格朗日插值法
f_lag([4, 5])
```

```python
plt.scatter(x, y) # 绘制散点图
plt.scatter([4, 5], model([4, 5])) # 绘制线性插值得到4,5对应值的点，红色
plt.scatter([4, 5], f_lag([4, 5])) # 绘制拉格朗日插值法得到的4,5对应的值的点，绿色
plt.plot(x, model(x), 'r-') # 绘制线性插值后的拟合的线，红色
plt.plot(x, f_lag(x), 'g-') # 绘制拉格朗日插值后拟合的线，绿色
```

```python
from scipy.interpolate import spline, BSpline
y_bs = BSpline(x, y, k=1) # 使用样条插值
y_bs([4, 5]) # 预测4,5的值
```

# 标准化数据

## 离差标准化数据

```python
import pandas as pd
data = pd.read_csv('../data/detail.csv', encoding='gbk')
data.head()
```

```python
data[['counts', 'amounts']]
```

```python
def MinMaxScale(data): # 离差标准化
    return (data - data.min()) / (data.max() - data.min())
a = MinMaxScale(data['counts']) # 当数据和最小值相等时，通过离差标准化数据就变为0了
b = MinMaxScale(data['amounts']) # 范围是0-1之间
pd.concat([a, b], axis=1) # 纵向合并两个表
```

## 标准差标准化数据

```python
def StandScale(data): # 标准差标准化
    return (data - data.mean()) / data.std()
a = StandScale(data['counts']) 
b = StandScale(data['amounts']) # 取值不限于0-1  同时离差标准化和标准差标准化都不会改变数据分布的情况
pd.concat([a, b], axis=1) # 纵向合并两个表
```

## 小数定标标准化数据

```python
import numpy as np
def DecimalScale(data):
    return data / 10**(np.ceil(np.log10(data.abs().max()))) # np.ceil()向上取整 abs()绝对值
a = DecimalScale(data['counts']) 
b = DecimalScale(data['amounts']) # -1 - 1 内
pd.concat([a, b], axis=1) # 纵向合并两个表
```

## 离散化处理

```python
# 等宽法 要求比较高，数据分布最好均匀
pd.cut(data['amounts'], bins=5)
pd.cut(data['amounts'], bins=5).value_counts()
```

```python
# 等频法 
def samefreq(data, k):
    w = data.quantile(np.arange(0, 1+1/k, 1/k)) # 切分数据的区间
    return pd.cut(data, w) # w 对应 bins
samefreq(data['amounts'], k=5)
samefreq(data['amounts'], k=5).value_counts() # 这样通过等频法就比等差法在各区间的分配就规律的多
```
