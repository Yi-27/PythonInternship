```python
import numpy as np
import pandas as pd

# 1. 读取数据
job_info = pd.read_csv("../data/job_info.csv", encoding="gbk")
job_info.head()
```

```python
# 2. 列名重命名
job_info.rename(columns={"字节跳动": "公司", 
                         "数据产品经理": "岗位", 
                         "北京": "工作地点", 
                         "2-3.5万/月": "工资", 
                         "09-03": "发布日期"}, inplace=True)
job_info.columns
```

```python
# 3. 哪个岗位招聘需求最多
print(job_info['岗位'].describe()) # 也可以通过describe()查看哪个岗位是最多的
print(job_info['岗位'].value_counts()) # 通过value_counts()查看岗位各元素的统计个数
job_info['岗位'].value_counts().index[0] # 提取最多的岗位
```

```python
# 4. 取出9月3日发布的招聘信息
print(job_info['发布日期'].value_counts()) # 先查看有哪些日期的数据,并统计看各日期的数据各有多少
print(job_info.loc[job_info['发布日期']=='09-03', :]) # 提取出9月三日的数据
job_info.loc[job_info['发布日期']=='09-03', :].shape # 可以看出数据共5314条，与提取前一致，故提取成功
```

```python
# 5. 找出工作地点在深圳的数据分析师招聘信息
gzdd = job_info.loc[job_info["工作地点"]=="深圳", :] # 先提取出深圳的数据
print(gzdd)
gzdd.loc[gzdd["岗位"]=="数据分析师", :] # 在提取出数据分析师的招聘信息
```

```python
# 6. 取出每条招聘信息中的最低工资与最高工资，单位为“元/月”
#    若招聘信息中无工资数据则无需处理
#    通过检查数据发现工资的单位，并不是固定的，有好几种格式
#    千/月  万/月  万/年  这三个算是比较正常的
#    元/天  元/小时  1.5千以下/月  这三个是不正常的只有一个工资数据，故最低最高工资设为一样
job_info["最低工资"] = np.nan # 先添加这两列
job_info["最高工资"] = np.nan

for i in range(job_info.shape[0]):
    if pd.isna(job_info.loc[i, "工资"]) == True:
        continue
    try:
        if "以下" in job_info.loc[i, "工资"]: # 经检查只有1.5千以下/月这一种特殊情况，这里最低最高也都设为一样
            min = max = job_info.loc[i, "工资"][:-5]
            money_unit, time_unit = job_info.loc[i, "工资"][-5], job_info.loc[i, "工资"][-1]
        elif "小时" in job_info.loc[i, "工资"]: # 元/小时的情况
            min = max = job_info.loc[i, "工资"][:-4]
            money_unit, time_unit = job_info.loc[i, "工资"][-4:].split('/') # 提取出工资单位和时间单位
        else:
            money_list = job_info.loc[i, "工资"][:-3].split('-')
            if len(money_list) == 2: # 提取出最低工资和最高工资
                min, max =  money_list[0], money_list[1]
            else:
                min = max = money_list[0] # 元/天的情况
            money_unit, time_unit = job_info.loc[i, "工资"][-3:].split('/') # 提取出工资单位和时间单位
    except Exception as e:
        print(e)
        print(job_info.loc[i, "工资"], i)
    if time_unit == "小时": # 每天算工作 12 个小时
        job_info.loc[i, "最低工资"] = str(int(float(min) * 12 * 30)) + "元/月"
        job_info.loc[i, "最高工资"] = str(int(float(max) * 12 * 30)) + "元/月"
    elif time_unit == "天": # 每个月都按30天算
        job_info.loc[i, "最低工资"] = str(int(float(min) * 30)) + "元/月"
        job_info.loc[i, "最高工资"] = str(int(float(max) * 30)) + "元/月"
    elif time_unit == "月":
        if money_unit == "千":
            job_info.loc[i, "最低工资"] = str(int(float(min) * 1000)) + "元/月"
            job_info.loc[i, "最高工资"] = str(int(float(max) * 1000)) + "元/月"
        elif money_unit == "万":
            job_info.loc[i, "最低工资"] = str(int(float(min) * 10000)) + "元/月"
            job_info.loc[i, "最高工资"] = str(int(float(max) * 10000)) + "元/月"
        else:
            print(money_unit) # 如果存在其他工资单位，先打印出来
    elif time_unit == "年":
        if money_unit == "万":
            job_info.loc[i, "最低工资"] = str(int(float(min) * 10000 / 12)) + "元/月"
            job_info.loc[i, "最高工资"] = str(int(float(max) * 10000 / 12)) + "元/月"
        else:
            print(money_unit) # 如果存在其他工资单位，先打印出来
    else:
        print(time_unit, i) # 如果存在其他时间单位，先打印出来
job_info
# 这里存在一些问题
# ① 可以用正则表达式来提取元素
# ② 代码执行速度过慢
# ③ 代码过于冗余，复用性太差，也不美观
# 这些问题是之后要注意的，虽然花点时间能做到，但是今天太忙了，之后抽时间再考虑一下 2020/7/8 11:23
```
