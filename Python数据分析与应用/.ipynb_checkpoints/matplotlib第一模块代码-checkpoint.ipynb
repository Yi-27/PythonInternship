```python
import numpy as np
import matplotlib.pyplot as plt

plt.rcParams['font.sans-serif'] = ['SimHei'] # 解决中文乱码
plt.rcParams['axes.unicode_minus'] = False # 解决符号乱码

# 散点图
x = np.linspace(0, 2*np.pi, 100) # 生成0 - 2π之间 100个数构成的等差数列
y = np.sin(x) + np.random.random(len(x)) # 生成同样个数的y

plt.figure(figsize=(8,5)) # 设置画布大小
plt.title("这是sin散点图")
plt.scatter(x, y)
plt.show()
```

```python
# 折线图
x = np.linspace(0, 2*np.pi, 100)
y = np.sin(x) + np.random.random(100)

plt.figure(dpi=120) # 设置分辨率
plt.plot(x, y, 'r--')
plt.plot(x, np.sin(x), 'g') # 光滑sin(x)曲线
plt.legend(['折线', 'sin曲线'])
plt.show()
```

```python
# 柱状图
x = range(10) # 随机10个数
np.random.seed(123) # 设置随机种子
y1 = np.random.random(10)
y2 = np.random.random(10)

plt.bar(x, y1, facecolor='r')
plt.bar(x, -y2, facecolor='g') # 由于y2是ndarray类型的数据，所以可以直接 -y2批量修改数据

# 设置文字说明
for i, j in zip(x, y1): # zip() 返回的是一个对象。如需展示列表，需手动 list() 转换。
    plt.text(i, j+0.05, '%.2f'%j, ha='center', va='bottom') # ha文字水平的位置，va文字垂直的位置

# +0.05 和 -0.05 是为了让文本比柱状图高一点，好直观显示而已
for i, j in zip(x, y2):
    plt.text(i, -j-0.05, '%.2f'%j, ha='center', va='top') # ha文字水平的位置，va文字垂直的位置

plt.ylim(-1.2, 1.2) # 设置Y轴的刻度范围
plt.title(r'$\sigma_i=15$') # TeX方程表达式 （数学公式）
plt.show()
```

```python
# 饼图
z = np.ones(10)

plt.figure(figsize=(5,5)) # autopct饼内文字的格式 explode突出显示饼 labeldistance表示label距离饼的距离
plt.pie(z, autopct='%.2f%%', explode=[0.1] + [0]*9, labels=list("ABCDEFGHIJ"), labeldistance=1.1) 
plt.show()
```
