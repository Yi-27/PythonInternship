```python
v = data.loc[data['order_id']=='458', :] # 提取出order_id为458的行所有列的数据
v
```

```python
v['order_id'] = '45800'# 更新order_id这一列的值
v  # 但只是对于v改动了，并没有改动源文件
```

```python
v['total_price'] = v['counts'] * v['amounts'] # 新增了一列
v.head(5)
```

```python
v.drop(labels='total_price', axis=1) # 删除上面添加的列，由于参数inplace默认为False，不会改动v的值，返回新的DataFrame
```

```python
v.drop(labels='total_price', axis=1, inplace=True) # 这样才会把v中真正的列给删除掉
v.columns
```

```python
pd.to_datetime(data['lock_time']) # 将lock_time这一列的时间字符串转换为标准时间类型
```

```python
pd.Timestamp.max # 查看时间戳的最大值
```

```python
data['use_start_time']
```

```python
pd.DatetimeIndex(data['use_start_time']) # 转换为DatetimeIndex格式 一组Timestamp构成的Index
```

```python
pd.PeriodIndex(data['use_start_time'], freq="S") # 指定时间间隔为 S秒  一组Period构成的Index
```
